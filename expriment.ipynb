Experiment to test on control information and relevant information

```python
from simple_gridworld import SimpleGridWorld
from mdp_graph import OptimalPolicyGraph
```

```python
# make H shape gridworld environment and mdp graph
env = SimpleGridWorld('maps/H.txt', make_random=False, random_traps=0,
                          agent_position=None, goal_position=(1, 8), max_steps=1024, trans_prob=1.0)
obs, info = env.reset()
mdp_graph = env.make_mdp_graph()
```

```python
# visualize the mdp graph
mdp_graph.visualize(figsize=(8, 5))
```

```python
# compute optimal policy and control information
optimal_policy_graph = OptimalPolicyGraph()
optimal_policy_graph.load_graph(mdp_graph)
optimal_policy_graph.value_iteration(0.999)
optimal_policy_graph.compute_optimal_policy(0.999)
optimal_policy_graph.control_info_iteration(0.999)
```

```python
# visualize control information and policy distribution
optimal_policy_graph.visualize_policy_and_control_info(figsize=(8, 5))
optimal_policy_graph.draw_action_distribution(figsize=(8, 2))
```

```python
# visualize control information and policy distribution with highlighted states and compute corresponding relevant information
highlighted_states = {(2, i) for i in range(2, 8)}
relevant_info = optimal_policy_graph.compute_policy_relevant_info(state_set=highlighted_states)
optimal_policy_graph.visualize_policy_and_control_info(title=f"Relevant Information: {relevant_info}", highlight_states=highlighted_states, figsize=(8, 5))

highlighted_states = {(3, i) for i in range(2, 8)}
relevant_info = optimal_policy_graph.compute_policy_relevant_info(state_set=highlighted_states)
optimal_policy_graph.visualize_policy_and_control_info(title=f"Relevant Information: {relevant_info}", highlight_states=highlighted_states, figsize=(8, 5))

highlighted_states = {(2, i) for i in range(2, 8)}.union({(3, i) for i in range(2, 8)})
relevant_info = optimal_policy_graph.compute_policy_relevant_info(state_set=highlighted_states)
optimal_policy_graph.visualize_policy_and_control_info(title=f"Relevant Information: {relevant_info}", highlight_states=highlighted_states, figsize=(8, 5))
```
